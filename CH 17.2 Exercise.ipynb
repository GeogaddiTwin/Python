```python
# view title tables contents
pd.read_sql('SELECT * FROM titles', connection)
```

```python
# connect author with isbn abd view first 5 rows
df = pd.read_sql('SELECT * FROM author_ISBN', connection)
df.head()
```

```python
# return first and last from authors table
pd.read_sql('SELECT first, last FROM authors', connection)
```

```python
# select title, edition and copyright for books with copyright years greater than 2016
pd.read_sql("""SELECT title, edition, copyright
            FROM titles
            WHERE copyright > '2016'""", connection)
```

```python
# find authors whose last name starts with letter d
pd.read_sql("""SELECT id, first, last
        FROM authors
        WHERE last LIKE 'D%'""",
        connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last
        FROM authors
        WHERE first LIKE '_b%'""",
        connection, index_col=['id'])
```

```python
# sort titels in ascending order
pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
        connection)
```

```python
# sort authors name by lasat name, also sort first name for authors who have the same last name
pd.read_sql("""SELECT id, first, last
            FROM authors
            ORDER BY last, first""",
            connection, index_col=['id'])
```

```python
# sort the authors in descending order by
#last name and ascending order by first name for any authors who have the same last name
pd.read_sql("""SELECT id, first, last
            FROM authors
            ORDER BY last DESC, first ASC""",
            connection, index_col=['id'])
```

```python
# get the isbn, title, edition and copyright of each book in the titles table that has a title ending with
#'How to Program' and sort them in ascending order by title.
pd.read_sql("""SELECT isbn, title, edition, copyright
                FROM titles
                WHERE title LIKE '%How to Program'
                ORDER BY title""", connection)
```

```python
# use INNER JOIN to make a list of authers with the ISBN for books written by the author
pd.read_sql("""SELECT first, last, isbn
    FROM authors
    INNER JOIN author_ISBN
    ON authors.id = author_ISBN.id
    ORDER BY last, first""", connection).head()
```

```python
cursor = connection.cursor()

# Use INSERT INTO to insert a author into the authors table
cursor = cursor.execute("""INSERT INTO authors (first, last)
                        VALUES ('Sue', 'Red')""")

# Query authors table to check auto generated id numbers
pd.read_sql('SELECT id, first, last FROM authors',
    connection, index_col=['id'])

cursor.close()
```

```python
# update author sue's lastname to Black

cursor = cursor.execute("""UPDATE authors SET last='Black'
                        WHERE last='Red' AND first='Sue'""")
# see how many rows were updated
cursor.rowcount
```
