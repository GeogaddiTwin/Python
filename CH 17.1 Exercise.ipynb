```python
'''
Name: Daniel Bartholomew
Date: 12/8/22
Descr: Program that covers sorting the book database using sqlite3
'''

import sqlite3
connection = sqlite3.connect('books.db')

import pandas as pd
pd.options.display.max_columns = 10
```

# Part A

```python
# sort authors last name in descending order
pd.read_sql("""SELECT id, last
            FROM authors
            ORDER BY last DESC""",
            connection, index_col=['id'])
```

# Part B

```python
# sort book titels in ascending order
pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
        connection)
```

# Part C

```python
# c) Use an INNER JOIN to select all the books for a specific author. Include the title
# copyright year and ISBN. Order the information alphabetically by title.

pd.read_sql("""SELECT first, last, title
    FROM author_ISBN INNER JOIN authors
    ON authors.first = authors.first
    INNER JOIN titles
    ON titles.title = titles.title
    ORDER BY title""",connection)
```
